```python
CAT=[]
CON=[]
for i in A.columns:
    if (A[i].dtypes=="object"):
        CAT.append(i)
    else:
        CON.append(i)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
x=len(CAT)
y=len(CON)
```

```python
x
```

```python
y
```

```python
m=(x+1)/2
n=(y+1)/2
```

```python
i=1
plt.figure(figsize=(30,23))
for q in CAT:
           plt.subplot(m,2,i)
           sb.countplot(A[q])
           i+=1
```

```python
i=1
plt.figure(figsize=(30,23))
for q in CON:
           plt.subplot(n,2,i)
           sb.distplot(A[q])
           i+=1
```
